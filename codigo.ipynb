```python
# Passo a passo do projeto
# Passo 1: Importar base de dados
import pandas as pd 

tabela = pd.read_csv("cancelamentos_sample.csv")
display(tabela)
# Passo 2: Visualizar a base de dados
    # Enteder quais informações você tem disponível
    # Identificar os erros da base de dados
tabela = tabela.drop(columns="CustomerID")
display(tabela)
```

```python
#Passo 3: Tratamento de Dados = Corrigir os erros da base de dados 
# valores vazios na base de dados
# formato das informações está correto

# Nan = not a number
 
tabela = tabela.dropna()
display(tabela.info())
#Passo 4: Analisar o cancelamento dos clientes
#Passo 5: Analisar a causa do cancelamento dos clientes(com as informações do cliente impactam no cancelamento)
```

```python
#Passo 4: Analisar o cancelamento dos clientes
display(tabela["cancelou"].value_counts())
display(tabela["cancelou"].value_counts(normalize=True))
```

```python
#Passo 5: Analisar a causa do cancelamento dos clientes(com as informações do cliente impactam no cancelamento)

# importar a biblioteca de gráficos
import plotly.express as px 

# criando grafico
grafico = px.histogram(tabela, x= "duracao_contrato")

# exibir o grafico
grafico.show()
```
